## Análise 1: Quantos clientes existem em cada grupo?

```python
df_contagem = df_clientes.groupby('grupo').size().reset_index(name='qtd_pessoas')
```

```python
# --- CÓDIGO INALTERADO (INÍCIO) ---
# Criamos uma cópia para não alterar o DataFrame original
df_agregado = df_cluster.copy()

# Renomeamos 'cluster' para 'grupo' e 'Qtd' para 'qtd_pessoas'
df_agregado.rename(columns={'cluster': 'grupo', 'Qtd': 'qtd_pessoas'}, inplace=True)

# Criamos a coluna 'categoria'
df_agregado['categoria'] = df_agregado['grupo'].apply(lambda x: 'Pessoa' if 'Pessoa' in x else 'Empresa')

# 2. Separação em DataFrames 'Pessoa' e 'Empresa'
df_pessoa = df_agregado[df_agregado['categoria'] == 'Pessoa'].sort_values('qtd_pessoas', ascending=False)
df_empresa = df_agregado[df_agregado['categoria'] == 'Empresa'].sort_values('qtd_pessoas', ascending=False)
# --- CÓDIGO INALTERADO (FIM) ---


# --- FUNÇÃO CORRIGIDA ---
def formatar_numero_compacto(n):
    """Formata um número para um formato compacto (e.g., 1.2k, 15.1M)."""
    # ALTERAÇÃO: Garante que números abaixo de 1000 sejam exibidos como inteiros.
    if abs(n) < 1000:
        return f'{n:.0f}'

    # Milhares (k)
    elif abs(n) < 1_000_000:
        num_formatado = f'{n / 1000:.1f}'
        return num_formatado.replace('.0', '') + 'k'

    # Milhões (M)
    elif abs(n) < 1_000_000_000:
        num_formatado = f'{n / 1_000_000:.1f}'
        return num_formatado.replace('.0', '') + 'M'

    # Bilhões (G)
    else:
        num_formatado = f'{n / 1_000_000_000:.1f}'
        return num_formatado.replace('.0', '') + 'G'


# 3. Geração dos Gráficos (código inalterado a partir daqui)
fig, axes = plt.subplots(1, 2, figsize=(16, 7))
fig.suptitle('Distribuição de Clientes por Categoria', fontsize=18)

# --- Gráfico da Esquerda (Pessoa) ---
bar_plot_pessoa = sns.barplot(ax=axes[0], x='qtd_pessoas', y='grupo', data=df_pessoa, palette='Blues_d')
axes[0].set_title('Grupos "Pessoa"', fontsize=14)
axes[0].set_xlabel('Quantidade de Pessoas', fontsize=12)
axes[0].set_ylabel('')

for bar in bar_plot_pessoa.patches:
    bar_width = bar.get_width()
    y_pos = bar.get_y() + bar.get_height() / 2
    label_text = formatar_numero_compacto(bar_width)
    x_pos = bar_width * 0.98
    axes[0].text(x_pos, y_pos, label_text,
                 ha='right', va='center', color='white',
                 fontsize=11, fontweight='bold')

# --- Gráfico da Direita (Empresa) ---
bar_plot_empresa = sns.barplot(ax=axes[1], x='qtd_pessoas', y='grupo', data=df_empresa, palette='Greens_d')
axes[1].set_title('Grupos "Empresa"', fontsize=14)
axes[1].set_xlabel('Quantidade de Empresas', fontsize=12)
axes[1].set_ylabel('')

for bar in bar_plot_empresa.patches:
    bar_width = bar.get_width()
    y_pos = bar.get_y() + bar.get_height() / 2
    label_text = formatar_numero_compacto(bar_width)
    x_pos = bar_width * 0.98
    axes[1].text(x_pos, y_pos, label_text,
                 ha='right', va='center', color='white',
                 fontsize=11, fontweight='bold')

# --- CÓDIGO FINAL INALTERADO ---
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

nome_arquivo = 'Distribuição de Clientes por Categoria.png'
caminho_completo = os.path.join(PASTA_GRAFICOS, 'geral/' + nome_arquivo)
plt.savefig(caminho_completo, bbox_inches='tight')

plt.show()
plt.close()
```

```python
# --- Processamento (inalterado) ---
df_agregado = df_cluster.copy()
df_agregado.rename(columns={'cluster': 'grupo'}, inplace=True)
df_agregado['categoria'] = df_agregado['grupo'].apply(lambda x: 'Pessoa' if 'Pessoa' in x else 'Empresa')

# --- Funções de Formatação ---

def formatar_valor_medio(n):
    """Formatação para Recência, permitindo uma casa decimal."""
    if n == int(n):
        return f'{int(n)}'
    else:
        return f'{n:.1f}'

def formatar_valor_monetario(n):
    """Formatação monetária compacta para Gasto Médio."""
    if abs(n) < 1000:
        return f'R$ {n:.0f}'
    elif abs(n) < 1_000_000:
        num_formatado = f'{n / 1000:.1f}'.replace('.0', '')
        return f'R$ {num_formatado}k'
    else:
        num_formatado = f'{n / 1_000_000:.1f}'.replace('.0', '')
        return f'R$ {num_formatado}M'

# NOVA FUNÇÃO para formatação de inteiros
def formatar_inteiro(n):
    """Formatação para exibir um número arredondado como inteiro."""
    return f'{n:.0f}'

# --- Função de Gráfico (inalterada) ---
def criar_grafico_rfm(df, coluna, titulo_geral, xlabel, nome_arquivo, func_formatacao):
    fig, axes = plt.subplots(1, 2, figsize=(16, 7))
    fig.suptitle(titulo_geral, fontsize=18)

    df_pessoa = df[df['categoria'] == 'Pessoa'].sort_values(coluna, ascending=False)
    df_empresa = df[df['categoria'] == 'Empresa'].sort_values(coluna, ascending=False)

    # --- Gráfico da Esquerda (Pessoa) ---
    bar_plot_pessoa = sns.barplot(ax=axes[0], x=coluna, y='grupo', data=df_pessoa, palette='Blues_d')
    axes[0].set_title('Grupos "Pessoa"', fontsize=14)
    axes[0].set_xlabel(xlabel, fontsize=12)
    axes[0].set_ylabel('')
    max_width_pessoa = df_pessoa[coluna].max() if not df_pessoa.empty else 0
    axes[0].set_xlim(right=max_width_pessoa * 1.15)

    for bar in bar_plot_pessoa.patches:
        bar_width = bar.get_width()
        y_pos = bar.get_y() + bar.get_height() / 2
        label_text = func_formatacao(bar_width)
        if bar_width < max_width_pessoa * 0.10:
            x_pos = bar_width + (max_width_pessoa * 0.01)
            axes[0].text(x_pos, y_pos, label_text, ha='left', va='center', color='black', fontsize=10)
        else:
            x_pos = bar_width * 0.98
            axes[0].text(x_pos, y_pos, label_text, ha='right', va='center', color='white', fontsize=11, fontweight='bold')

    # --- Gráfico da Direita (Empresa) ---
    bar_plot_empresa = sns.barplot(ax=axes[1], x=coluna, y='grupo', data=df_empresa, palette='Greens_d')
    axes[1].set_title('Grupos "Empresa"', fontsize=14)
    axes[1].set_xlabel(xlabel, fontsize=12)
    axes[1].set_ylabel('')
    max_width_empresa = df_empresa[coluna].max() if not df_empresa.empty else 0
    axes[1].set_xlim(right=max_width_empresa * 1.10)

    for bar in bar_plot_empresa.patches:
        bar_width = bar.get_width()
        y_pos = bar.get_y() + bar.get_height() / 2
        label_text = func_formatacao(bar_width)
        if bar_width < max_width_empresa * 0.10:
            x_pos = bar_width + (max_width_empresa * 0.01)
            axes[1].text(x_pos, y_pos, label_text, ha='left', va='center', color='black', fontsize=10)
        else:
            x_pos = bar_width * 0.98
            axes[1].text(x_pos, y_pos, label_text, ha='right', va='center', color='white', fontsize=11, fontweight='bold')

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    caminho_completo = os.path.join(PASTA_GRAFICOS, 'geral', nome_arquivo)
    plt.savefig(caminho_completo, bbox_inches='tight')
    plt.show()
    plt.close()

# --- Geração dos 3 Gráficos com a formatação correta para cada um ---

# 1. Gráfico de Recência Média (permite decimal)
criar_grafico_rfm(
    df=df_agregado, coluna='recency_mean',
    titulo_geral='Recência Média por Grupo', xlabel='Recência Média (dias)',
    nome_arquivo='Recencia_Media_por_Grupo.png',
    func_formatacao=formatar_valor_medio
)

# 2. Gráfico de Frequência Média (usa a NOVA formatação de inteiro)
criar_grafico_rfm(
    df=df_agregado, coluna='frequency_mean',
    titulo_geral='Frequência Média por Grupo', xlabel='Frequência Média de Compras',
    nome_arquivo='Frequencia_Media_por_Grupo.png',
    func_formatacao=formatar_inteiro
)

# 3. Gráfico de Gasto Médio Total (usa a formatação monetária)
criar_grafico_rfm(
    df=df_agregado, coluna='monetary_mean',
    titulo_geral='Gasto Médio Total por Grupo', xlabel='Gasto Médio Total (R$)',
    nome_arquivo='Gasto_Medio_Total_por_Grupo.png',
    func_formatacao=formatar_valor_monetario
)

print("Gráficos gerados com a formatação de frequência atualizada!")
```

## Análise 2: Quanto que fatura por dia da semana?

```python
# --- Preparação dos Dados ---
df_vendas['date_purchase'] = pd.to_datetime(df_vendas['date_purchase'])

dias_semana_map = {
    'Monday': '2-Segunda', 'Tuesday': '3-Terça', 'Wednesday': '4-Quarta',
    'Thursday': '5-Quinta', 'Friday': '6-Sexta', 'Saturday': '7-Sábado', 'Sunday': '1-Domingo'
}
df_vendas['dia_da_semana'] = df_vendas['date_purchase'].dt.day_name().map(dias_semana_map)

# --- Agrupamento dos Dados  ---
vendas_por_dia = df_vendas.groupby('dia_da_semana')['gmv_success'].sum().sort_index()

# --- Geração do Gráfico ---
plt.figure(figsize=(12, 7))
plot = sns.barplot(
    x=vendas_por_dia.index,
    y=vendas_por_dia.values,
    palette='plasma'
)

plt.title('Faturamento Total por Dia da Semana (2021 - 2024)', fontsize=18)
plt.xlabel('Dia da Semana', fontsize=12)
plt.ylabel('Faturamento Total (R$)', fontsize=12)
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, p: f'R$ {x:,.0f}'.replace(",", ".")))

# 1. Criamos a lista de textos com a formatação desejada (R$ 1.234.567)
labels_formatados = [f'R$ {val:_.0f}'.replace('_', '.') for val in vendas_por_dia.values]
# 2. Passamos a lista pronta para o parâmetro 'labels'
plot.bar_label(plot.containers[0], labels=labels_formatados, fontsize=10)

nome_arquivo = 'Faturamento_por_Dia_da_Semana.png'
caminho_completo = os.path.join(PASTA_GRAFICOS, 'geral/' + nome_arquivo)
plt.savefig(caminho_completo, bbox_inches='tight')

plt.show()
plt.close()
```

## Análise 3: Quais são os horários que os clientes mais compram durante a semana?

```python
# --- Preparação dos Dados ---
df_vendas['hora_da_compra'] = pd.to_datetime(df_vendas['time_purchase'], format='%H:%M:%S').dt.hour

# --- Agrupamento dos Dados ---
faturamento_por_hora = df_vendas.groupby('hora_da_compra')['gmv_success'].sum()

# --- Geração do Gráfico ---
plt.figure(figsize=(14, 7))
plot = sns.lineplot(
    x=faturamento_por_hora.index,
    y=faturamento_por_hora.values,
    marker='o',
    color='purple'
)
plt.title('Faturamento por Hora do Dia (2021 - 2024)', fontsize=18)
plt.xlabel('Hora do Dia (0-23h)', fontsize=12)
plt.ylabel('Faturamento Total (R$)', fontsize=12)
plt.xticks(range(0, 24)) # Garante que todas as horas apareçam no eixo X
plt.grid(True, linestyle='--', alpha=0.6)
# Formata o eixo Y para o padrão monetário
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, p: f'R$ {x:,.0f}'.replace(",", ".")))
nome_arquivo = 'Faturamento_por_Hora_do_Dia.png'
caminho_completo = os.path.join(PASTA_GRAFICOS, 'geral/' + nome_arquivo)
plt.savefig(caminho_completo, bbox_inches='tight')
plt.show()
plt.close()
```

## Análise 2.1: Quais são os horários que os clientes mais compram em um dia específico da semana?

```python
# --- Assumindo que 'df_vendas' já foi carregado e pré-processado como antes ---
if 'df_vendas' not in locals():
    print("DataFrame 'df_vendas' não encontrado. Carregue e prepare seus dados primeiro.")
else:
    # Agrupamos os dados de uma vez só para eficiência
    faturamento_por_dia_hora = df_vendas.groupby(['dia_da_semana', 'hora_da_compra'])['gmv_success'].sum().reset_index()

    # --- Geração do Gráfico (A Nova Abordagem) ---

    # 1. Configurar o Estilo e a Estrutura do Gráfico
    sns.set_style("white")
    plt.rcParams['font.family'] = 'sans-serif'
    fig, axes = plt.subplots(4, 2, figsize=(16, 20), sharex=True, sharey=True)
    axes = axes.flatten()
    fig.delaxes(axes[7]) # Remove o 8º eixo que não será usado

    # 2. Título Principal e Subtítulo (A História)
    fig.suptitle(
        'Padrão de Faturamento Horário Varia Durante a Semana',
        fontsize=24,
        fontweight='bold',
        ha='center',
        x=0.52
    )
    fig.text(0.52, 0.95, 'Faturamento total por hora do dia, destacando a diferença entre dias de semana e fins de semana (2021-2024)',
             fontsize=14, ha='center', color='#404040')

    # 3. Loop para Criar Cada Mini-Gráfico (Small Multiples)
    dias_ordenados = sorted(faturamento_por_dia_hora['dia_da_semana'].unique())

    for i, dia in enumerate(dias_ordenados):
        ax = axes[i]
        dados_dia = faturamento_por_dia_hora[faturamento_por_dia_hora['dia_da_semana'] == dia]

        # Define a cor estrategicamente: destaque para fim de semana
        cor_linha = '#005A9C' # Azul escuro para destaque
        cor_fundo = 'lightgray'
        cor_pico = 'red'

        # Plota todas as linhas em cinza claro para servirem de contexto
        for d_fundo in dias_ordenados:
            dados_fundo = faturamento_por_dia_hora[faturamento_por_dia_hora['dia_da_semana'] == d_fundo]
            sns.lineplot(x='hora_da_compra', y='gmv_success', data=dados_fundo, ax=ax, color=cor_fundo, linewidth=1.5, alpha=0.5)

        # Plota a linha do dia atual em destaque por cima das outras
        sns.lineplot(x='hora_da_compra', y='gmv_success', data=dados_dia, ax=ax, color=cor_linha, linewidth=3.0)

        # Adiciona um ponto no pico de faturamento de cada dia
        pico = dados_dia.loc[dados_dia['gmv_success'].idxmax()]
        ax.scatter(pico['hora_da_compra'], pico['gmv_success'], color=cor_pico, s=100, zorder=5)
        
        # Anotação do pico
        ax.text(pico['hora_da_compra'], pico['gmv_success'] * 1.05, f"Pico: {pico['hora_da_compra']}h",
                ha='center', va='bottom', fontsize=10, fontweight='bold')

        # Títulos e formatação de cada mini-gráfico (reduzir clutter)
        ax.set_title(dia.split('-')[1], fontsize=16, loc='left', fontweight='bold', pad=10)
        ax.set_ylabel('')
        ax.set_xlabel('')
        sns.despine(ax=ax, left=True) # Remove todas as bordas e o eixo Y
        ax.grid(axis='x', linestyle='--', alpha=0.0) # Remove a grade vertical
        
        # Remove os ticks do eixo Y para um visual mais limpo
        ax.tick_params(axis='y', length=0)
        ax.set_yticklabels([])

    # 4. Rótulos Comuns e Layout Final
    fig.text(0.5, 0.08, 'Hora do Dia (0-23h)', ha='center', va='center', fontsize=14)
    plt.tight_layout(rect=[0, 0.1, 1, 0.93]) # Ajusta para caber títulos
    plt.savefig(os.path.join(PASTA_GRAFICOS, 'geral/Faturamento_Horario_por_Dia_da_Semana_Small_Multiple.png'), bbox_inches='tight')
    plt.show()
```
